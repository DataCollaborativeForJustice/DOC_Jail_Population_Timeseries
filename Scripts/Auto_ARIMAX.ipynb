# Retrieve DOC Admission and Discharge Data from NYC Open Data Portal

### Socrata API urls:

* Inmate Admission: `https://data.cityofnewyork.us/resource/6teu-xtgp.json`

* Inmate Discharges: `https://data.cityofnewyork.us/resource/94ri-3ium.json`

Both of these datasets are unique on the inmate identification number `INMATEID`. We are going to count the number of inmates admitted and discharges daily from 2016 to May 1, 2023 using the SODA SQL language.

```python
admit_url = 'https://data.cityofnewyork.us/resource/6teu-xtgp.json'
dis_url = 'https://data.cityofnewyork.us/resource/94ri-3ium.json'

def get_data(url):
    query = (url+'?'
            "$select=*"
            "&$limit=500000")
    query = query.replace(" ", "%20")
    response = urllib.request.urlopen(query)
    data = json.loads(response.read())
    
    #store in dataframe
    df = pd.DataFrame(data,columns = data[0].keys())
    
    return df
```

```python
admissions = get_data(admit_url)
admissions.head()
```

```python
discharges = get_data(dis_url)
discharges.head()
```

```python
##CHANGE TIMESTAMP TO DATE

admissions['admitted_dt'] = pd.to_datetime(admissions['admitted_dt'])
discharges['discharged_dt'] = pd.to_datetime(discharges['discharged_dt'])

admissions['admitted_wk'] = admissions['admitted_dt'].dt.isocalendar().week
discharges['discharged_wk'] = discharges['discharged_dt'].dt.isocalendar().week

admissions['admitted_yr'] = admissions['admitted_dt'].dt.year
discharges['discharged_yr'] = discharges['discharged_dt'].dt.year

# admissions['Year-Mo'] = ['-'.join(i) for i in zip(admissions["admitted_yr"].map(str),admissions["admitted_mo"].map(str))]
# discharges['Year-Mo'] = ['-'.join(i) for i in zip(discharges["discharged_yr"].map(str),discharges["discharged_mo"].map(str))]
admissions.head()
```

```python
# Calculate the count of unique inmate IDs for each admitted date
daily_admissions = admissions.groupby(['admitted_yr','admitted_wk'])['inmateid'].nunique().reset_index(name='inmate_admissions')


# Pivot the dataframe to get gender counts
gender_counts = pd.pivot_table(admissions, index=['admitted_yr','admitted_wk'], columns='gender', values='inmateid', aggfunc='nunique')
gender_counts.columns = gender_counts.columns + '_admit'

# Pivot the dataframe to get inmate status code counts
status_counts = pd.pivot_table(admissions, index=['admitted_yr','admitted_wk'], columns='inmate_status_code', values='inmateid', aggfunc='nunique')
status_counts.columns = status_counts.columns + '_admit'

# Merge the three dataframes based on admitted_date
daily_admissions = daily_admissions.merge(gender_counts, on=['admitted_yr','admitted_wk'], how='left')
daily_admissions = daily_admissions.merge(status_counts, on=['admitted_yr','admitted_wk'], how='left').fillna(0)

daily_admissions.head()
```

```python
# Calculate the count of unique inmate IDs for each discharge date
daily_dis = discharges.groupby(['discharged_yr','discharged_wk'])['inmateid'].nunique().reset_index(name='inmate_discharge')


# Pivot the dataframe to get gender counts
gender_counts = pd.pivot_table(discharges, index=['discharged_yr','discharged_wk'], columns='gender', values='inmateid', aggfunc='nunique')
gender_counts.columns = gender_counts.columns + '_dis'

# Pivot the dataframe to get inmate status code counts
status_counts = pd.pivot_table(discharges, index=['discharged_yr','discharged_wk'], columns='inmate_status_code', values='inmateid', aggfunc='nunique')
status_counts.columns = status_counts.columns + '_dis'

# Merge the three dataframes based on admitted_date
daily_dis = daily_dis.merge(gender_counts, on=['discharged_yr','discharged_wk'], how='left')
daily_dis = daily_dis.merge(status_counts, on=['discharged_yr','discharged_wk'], how='left').fillna(0)

daily_dis.head()
```

```python
#merge
daily_series = daily_admissions.merge(daily_dis, left_on = ['admitted_yr','admitted_wk'],right_on = ['discharged_yr','discharged_wk'])
daily_series = daily_series.rename(columns = {'admitted_yr':'Year','admitted_wk':'Week'})
daily_series = daily_series.drop(columns = ['discharged_yr','discharged_wk'])
daily_series['Gain/Loss'] = daily_series['inmate_admissions'] - daily_series['inmate_discharge']
daily_series.head()
```

From: https://www.analyticsvidhya.com/blog/2021/11/basic-understanding-of-time-series-modelling-with-auto-arimax/

"We add some essential features like lag values of available numeric features widely accepted for time series problems. Considering we need to foretell the stock price for a day, we cannot use the feature values of the same day since they will be unavailable at actual inference time. We require to use statistics like the mean, the standard deviation of their lagged values. The three sets of lagged values are used, one previous day, one looking back seven days and another looking back 30 days as a proxy for last week and last month metrics."

I will try to do the same with inmate gender and admission and discharge inmate status code variables

```python
from statsmodels.tsa.stattools import acf, pacf,adfuller, kpss,range_unit_root_test # BDS for inped
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

#plot autcorrelation for lags up to a year
plt.figure(figsize = (8,15))
plot_acf(daily_series['Gain/Loss'], lags = 52)
plt.show()
```

```python
import statsmodels.api as sm

#ADF-test(Original-time-series)
res = sm.tsa.adfuller(daily_series['Gain/Loss'].dropna(),regression='ct')
print('p-value:{}'.format(res[1]))
```

```python

#ADF-test(Original-time-series)
res = sm.tsa.adfuller(daily_series['Gain/Loss'].diff().dropna(),regression='ct')
print('p-value:{}'.format(res[1]))
```

```python
df = daily_series.copy()

lag_features = ['F _admit', 'M _admit', 'CS_admit',
               'CSP_admit', 'DE_admit', 'DEP_admit', 'DNS_admit', 'DPV_admit',
               'SCO_admit', 'SSR_admit', 'F _dis', 'M _dis', 'CS_dis',
               'CSP_dis', 'DE_dis', 'DEP_dis', 'DNS_dis',
               'DPV_dis', 'SCO_dis', 'SSR_dis']
#windows are of lags
window1 = 1 #1 week
window2 = 2 #2 weeks
window3 = 12
window4 = 52 #yearly

df_rolled_1 = df[lag_features].rolling(window=window1, min_periods=0)
df_rolled_2 = df[lag_features].rolling(window=window2, min_periods=0)
df_rolled_3 = df[lag_features].rolling(window=window3, min_periods=0)
df_rolled_4 = df[lag_features].rolling(window=window4, min_periods=0)

df_mean_1 = df_rolled_1.mean().shift(1).reset_index().astype(float)
df_mean_2 = df_rolled_2.mean().shift(1).reset_index().astype(float)
df_mean_3 = df_rolled_3.mean().shift(1).reset_index().astype(float)
df_mean_4 = df_rolled_4.mean().shift(1).reset_index().astype(float)

df_std_1 = df_rolled_1.std().shift(1).reset_index().astype(float)
df_std_2 = df_rolled_2.std().shift(1).reset_index().astype(float)
df_std_3 = df_rolled_3.std().shift(1).reset_index().astype(float)
df_std_4 = df_rolled_4.std().shift(1).reset_index().astype(float)

mean_columns = []
std_columns = []

for i, feature in enumerate(lag_features):
    mean_columns.append(df_mean_1[feature])
    mean_columns.append(df_mean_2[feature])
    mean_columns.append(df_mean_3[feature])
    mean_columns.append(df_mean_4[feature])
    
    std_columns.append(df_std_1[feature])
    std_columns.append(df_std_2[feature])
    std_columns.append(df_std_3[feature])
    std_columns.append(df_std_4[feature])

mean_df = pd.concat(mean_columns, axis=1)
std_df = pd.concat(std_columns, axis=1)

new_columns = []
for i, feature in enumerate(lag_features):
    new_columns.append(pd.Series(mean_df.iloc[:, 3*i].values, name=f"{feature}_mean_lag{window1}"))
    new_columns.append(pd.Series(mean_df.iloc[:, 3*i+1].values, name=f"{feature}_mean_lag{window2}"))
    new_columns.append(pd.Series(mean_df.iloc[:, 3*i+2].values, name=f"{feature}_mean_lag{window3}"))
    new_columns.append(pd.Series(mean_df.iloc[:, 3*i+3].values, name=f"{feature}_mean_lag{window4}"))

    new_columns.append(pd.Series(std_df.iloc[:, 3*i].values, name=f"{feature}_std_lag{window1}"))
    new_columns.append(pd.Series(std_df.iloc[:, 3*i+1].values, name=f"{feature}_std_lag{window2}"))
    new_columns.append(pd.Series(std_df.iloc[:, 3*i+2].values, name=f"{feature}_std_lag{window3}"))
    new_columns.append(pd.Series(std_df.iloc[:, 3*i+3].values, name=f"{feature}_std_lag{window4}"))
    
df = pd.concat([df] + new_columns, axis=1)

df.set_index(['Year','Week'], drop=True, inplace=True)
df.fillna(0, inplace=True)
df.reset_index(inplace=True)

df.head()
```

```python
exogenous_features = df.columns.to_list()[len(list(daily_series.columns)):]
print(exogenous_features)
```

```python
df_train = df[df.Year < 2021]
df_valid = df[df.Year >= 2021]

model = pmdarima.arima.auto_arima(df_train['Gain/Loss'],d = 1,exogenous=df_train[exogenous_features], trace=True, error_action="ignore", suppress_warnings=True)
model.fit(df_train['Gain/Loss'], exogenous=df_train[exogenous_features])

forecast = model.predict(n_periods=len(df_valid), exogenous=df_valid[exogenous_features])
df_valid["Forecast_ARIMAX"] = forecast
```
