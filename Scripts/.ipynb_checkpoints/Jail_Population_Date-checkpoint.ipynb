```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
from scipy.stats import norm

import urllib.request, json
import requests
import urllib.parse
import datetime
```

```python
#import inmate admission and discharge data
#last downloaded from AWS 7/18 AM
inmate_data = pd.read_csv('../Data/merged_socrata_today.csv')
inmate_data.head()
```

```python
inmate_data.dtypes
```

```python
inmate_data.columns
```

```python
#set variable types
inmate_data['admitted_dt'] = pd.to_datetime(inmate_data['admitted_dt'])
inmate_data['date_exited'] = pd.to_datetime(inmate_data['date_exited'],format='mixed')
inmate_data[['bradh', 'inmate_status_code', 'sealed',
             'srg_flg', 'infraction', 'race', 'gender',
             'top_charge','custody_level']] = inmate_data[['bradh', 'inmate_status_code', 'sealed',
                                                             'srg_flg', 'infraction', 'race', 'gender',
                                                             'top_charge','custody_level']].astype(str) 
inmate_data['age'] = inmate_data['age'].fillna(0).astype(int)
```

```python
print(inmate_data['admitted_dt'].min(),inmate_data['admitted_dt'].max())
print(inmate_data['date_exited'].min(),inmate_data['date_exited'].max())
```

```python
#inmates cannot get admitted before discharge date. Let's filter out the rows with inmates that have incorrect dates
inmate_data.sort_values(by = 'admitted_dt').head(10)
```

```python
# Step 2: Create 'month' column
inmate_data['admission_month'] = inmate_data['admitted_dt'].dt.to_period('M')
inmate_data['discharge_month'] = inmate_data['date_exited'].dt.to_period('M')

# Step 3: Group and count unique inmateid per month
monthly_admissions = inmate_data.groupby('admission_month')['inmateid'].nunique()
monthly_discharges = inmate_data.groupby('discharge_month')['inmateid'].nunique()

# Step 4: Plot the series
plt.plot(monthly_admissions.index.astype(str), monthly_admissions.values, label='Monthly Admissions')
plt.plot(monthly_discharges.index.astype(str), monthly_discharges.values, label='Monthly Discharges')

plt.xlabel('Month')
plt.ylabel('Number of Unique Inmates')
plt.title('Monthly Admissions and Discharges')
plt.legend()
plt.show()
```

Our data source began collecting data on daily jail population on 07/27/2021 and was last updated 07/18/2023 (today). We will create a weekly & monthly jail population timeseries by counting the number of unique inmate ids are currently in custody on any given week or month.

```python
start_date = '2021-07-27'
end_date = '2023-07-18'

# daily dates
dates = pd.date_range(start=start_date, end=end_date, freq='D')
```

```python
print(dates)
```

```python
dates[0]
```

```python
# Step 2: Create 'month' column
inmate_data['admitted_dt'] = inmate_data['admitted_dt'].dt.to_period('D').dt.to_timestamp()
inmate_data['date_exited'] = inmate_data['date_exited'].dt.to_period('D').dt.to_timestamp()
```

```python
inmate_data['admitted_dt'].iloc[0]
```

```python
daily_pop = []
for day in dates:
    daily_pop.append(inmate_data[(inmate_data['admitted_dt'] <= day) &
                                 ((inmate_data['date_exited'] > day)|(inmate_data['date_exited'].isna()))]['inmateid'].nunique()) 
```

```python
plt.plot(dates,daily_pop)
plt.xticks(rotation=45)  # Rotates x-axis tick labels for better readability
plt.show()
```

```python
len(inmate_data[inmate_data['date_exited'].isna()])
```

```python
daily_pop[0]
```

This method shows the daily jail population on 07/27/2021 was 4,021 and as of today it is 6,211. Let's fact check these numbers with Vera's Jaily Population Dashboard

Vera states the population on 07/27/2021 to be 5,900 and 6,211 on 07/17/2021.
